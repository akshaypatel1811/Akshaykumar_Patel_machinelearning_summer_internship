```python
import pandas as pd
import matplotlib.pyplot as plt
```

For sample province I have taken Ontario 

```python
df= pd.read_excel('Akshaykumar_Patel_machinelearning_by provinces pivoted by age.xlsx',skiprows = 6)
new_df=df.fillna(method="ffill")                                       # used fill forward method to fill the NA data
df_ontario= new_df[new_df.Geography2 == 'Ontario']                     # to get the data related to perticular province only
df_ontario.reset_index(inplace=True)                                   
df_ontario                                                             # to show how the data look like   
```

```python
plt.plot(df_ontario['Reference period'],df_ontario['15 to 24 years'],label='15 to 24 years')
plt.plot(df_ontario['Reference period'],df_ontario['25 to 44 years'],label='25 to 44 years')
plt.plot(df_ontario['Reference period'],df_ontario['45 to 64 years'],label='45 to 64 years')
plt.plot(df_ontario['Reference period'],df_ontario['65 to 69 years'],label='65 to 69 years')
plt.plot(df_ontario['Reference period'],df_ontario['70 years and over'],label='70 years and over')
plt.xlabel('Ontario')
plt.ylabel('Employment rate(%)')
plt.title('Employment rate of Ontario by age')
plt.legend(loc='best')
plt.ylim(0,140)
plt.savefig('Ontario employment rate by age over last 5 years.jpg')
```

## Conclusion

1. Employment is high in age group of 25 to 44 years as compair to other age group as it is the obvious working class people.
2. Employment rate trend is same in every year and it is steady.
3. Employment rate increase in summer time for age group of 15 to 24 years the reason could be summer vacation for high            school student which allow them to work.
