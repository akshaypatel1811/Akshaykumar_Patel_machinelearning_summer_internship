```python
import pandas as pd
import matplotlib.pyplot as plt
```

For sample province I have taken British Columbia

```python
df= pd.read_excel('AKshaykumar_Patel_machinelearning_by provinces pivoted by industry.xlsx',skiprows = 8)
new_df=df.fillna(method="ffill")                                       # used fill forward method to fill the NA data
df_British_Columbia= new_df[new_df.Geography2 == 'British Columbia']   # to get the data related to perticular province only
df_British_Columbia.set_index('North American Industry Classification System (NAICS)3',inplace = True) # Prepare dataframe for 
df_British_Columbia.drop(['Geography2'],axis = 1, inplace = True)      # bar chart
df_British_Columbia.replace('..',0,inplace = True)
df_British_Columbia = df_British_Columbia.apply(pd.to_numeric)         # to convert all obj type to float64
df_British_Columbia                                                    # to show how the data look like 
```

I have made Bar chart

```python
df_British_Columbia.plot.bar(figsize=(20,10),title= 'British columbia Employment by types of industries')
plt.ylabel('number of people (*1000)')
plt.savefig('British columbia Employment by types of industries.jpg')
```

## Conclusion

1. Most of the people are employed in Service-producing sector.
2. As the year passes number of people employed increase in most of the industries which is a good sign for the Province.
3. Agriculture,oil & gas, Forestry, Fishing, and Utilities are the sectors where less number of people are employed as compair    to other industry sector.
