In this notebook I have done Visualization for employment rate of pericular province as a sample ( here I have taken Alberta ) with respect to sex to see how it changes over time. Important notes are mention along with the code itself.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

For sample I have chose Alberta as a province

```python
df=pd.read_excel('Employment by provinces pivoted by sex.xlsx',skiprows=6)
new_df=df.fillna(method="ffill")                                       # used fill forward method to fill the NA data
df_alberta=new_df[new_df.Geography2=='Alberta']                        # to get the data related to perticular province only
df_alberta.reset_index(inplace=True)                                   # to reset index
df_alberta                                                             # to show how the data look like
```

Plotting of employment rate in Alberta

```python
plt.plot(df_alberta['Reference period'],df_alberta['Males'],label='Male')
plt.plot(df_alberta['Reference period'],df_alberta['Females'],label='Female')
plt.plot(df_alberta['Reference period'],df_alberta['Both sexes'],label='Both')
plt.xlabel('Albeta')
plt.ylabel('Employment rate(%)')
plt.title('Employment rate of Albeta by sex')
plt.legend(loc='best')
plt.ylim(57,80)
plt.savefig('Alberta employment rate by sex over last 5 years.jpg')
```

As we can see from this graph both male and female employment rate decrese over time and for female it is always lower than the man in every year
